```python
results_dir = "/s/red/b/nobackup/data/eggnog_cpm/training_files/exp_results/comparison/"

# bar plots comparison between rmpe and our
n_joints = len(pck_all_joints_0dot1)

def pck_bar_plots(rmpe_data, our_data, at):
    fig, ax = plt.subplots()
    fig.set_size_inches((10, 10))
    index = np.arange(n_joints)
    bar_width = 0.35

    opacity = 0.42
    error_config = {'ecolor': '0.3'}

    rects1 = plt.bar(index, rmpe_data, bar_width,
                     alpha=opacity,
                     color='r',
                     error_kw=error_config,
                     label='RMPE')

    rects2 = plt.bar(index + bar_width, our_data, bar_width,
                     alpha=opacity,
                     color='b',
                     error_kw=error_config,
                     label='Our')


    # plt.xlabel('Joints')
    plt.ylim((0.2,1.01))
    if at==0.1:
        out_at = "0dot1"
        plt.ylabel('PCK@0.1')
    if at==0.2:
        out_at = "0dot2"
        plt.ylabel('PCK@0.2')
    #plt.title('Comparison of RMPE tested with EGGNOG and our network PCK@' + str(at), fontsize = 15)
    ax.tick_params(axis='x', direction='out', length=10, width=2)
    fig.autofmt_xdate()

    plt.xticks(index + bar_width / 2, tuple(joints), rotation=30, fontsize = 20)
    plt.legend(loc='lower left')

    # plt.tight_layout()
    plt.show()
    
    fig.savefig(os.path.join(results_dir, npy_file_pck.split('.')[0] + "_rmpe_vs_our_@" + out_at + ".png"), bbox_inches='tight', pad_inches=0)
    fig.savefig(os.path.join(results_dir, npy_file_pck.split('.')[0] + "_rmpe_vs_our_@" + out_at + ".pdf"), format='pdf', dpi=300, bbox_inches='tight', pad_inches=0)
   

# @0.1
pck_bar_plots(pck_all_joints_0dot1_rmpe, pck_all_joints_0dot1, at=0.1)
# @0.2
pck_bar_plots(pck_all_joints_0dot2_rmpe, pck_all_joints_0dot2, at=0.2)
```
