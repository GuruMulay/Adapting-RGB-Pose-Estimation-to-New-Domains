```python
pck_at_idx = 5  # 5 => pck@0.1

def get_mean_pckY(pck_a, pck_b, pck_c, pck_d, pck_e):
    
    pcka = np.load(os.path.join(results_dir, pck_a))
    a = np.mean(pcka, axis=0)[pck_at_idx]  # from 10,26 to 1,26 to nth element of 1,26
    
    pckb = np.load(os.path.join(results_dir, pck_b))
    b = np.mean(pckb, axis=0)[pck_at_idx]
    
    pckc = np.load(os.path.join(results_dir, pck_c))
    c = np.mean(pckc, axis=0)[pck_at_idx]
    
    pckd = np.load(os.path.join(results_dir, pck_d))
    d = np.mean(pckd, axis=0)[pck_at_idx]
    
    pcke = np.load(os.path.join(results_dir, pck_e))
    e = np.mean(pcke, axis=0)[pck_at_idx]
    
    pck = np.mean(np.array([pcka, pckb, pckc, pckd, pcke]), axis=0)  # 10,26 matrix
    
    pck_at_n = np.array([a, b, c, d, e])  # pck at 0.1 for five indiviudal trials
    
    return pck, pck_at_n
    
```
